```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Read the CSV file into a DataFrame
data = pd.read_csv('cars fine.csv')

# Access the columns you want to plot
column_to_plot = 'Car'
column_to_plot2 = 'Speed in km/hr'
column_data = data[column_to_plot]
column_data2 = data[column_to_plot2]

# Define fines based on speed
fine = []
for i in column_data2:
    if i > 60 and i <= 70:
        fine.append(500)
    elif i > 70 and i <= 80:
        fine.append(750)
    elif i > 80:
        fine.append(1000)
    else:
        fine.append(0)  # Default value if speed does not meet any condition

# Plot the fines against speed

plt.title('Fine vs Speed')
plt.xlabel('Speed in km/hr')
plt.ylabel('Fine')
plt.grid(True,c='#0ABAB5')
plt.scatter(fine, column_data2,marker='*')  # Corrected order of arguments
fig2 = plt.figure()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Plot 1
plt.figure()
plt.plot(x, y1)
plt.title('Plot 1: Sine Wave')

# Plot 2
plt.figure()
plt.plot(x, y2)
plt.title('Plot 2: Cosine Wave')

# Plot 3
plt.figure()
plt.scatter(x, y1, color='red', label='Sine')
plt.scatter(x, y2, color='blue', label='Cosine')
plt.legend()
plt.title('Plot 3: Scatter Plot of Sine and Cosine')



# Display the plots
plt.show()

```

```python
import matplotlib.pyplot as plt

# Example data
categories = ['A', 'B', 'C', 'D']
values = [20, 35, 30, 15]

# Line plot
plt.figure(figsize=(8, 4))
plt.subplot(221)  # Subplot for line plot
plt.plot(categories, values, marker='o')
plt.title('Line Plot')

# Bar plot
plt.subplot(222)  # Subplot for bar plot
plt.bar(categories, values, color='skyblue')
plt.title('Bar Plot')

# Horizontal bar plot (barh)
plt.subplot(223)  # Subplot for horizontal bar plot
plt.barh(categories, values, color='salmon')
plt.title('Horizontal Bar Plot')

# Pie chart
plt.subplot(224)  # Subplot for pie chart
plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Pie Chart')

# Adjust layout
plt.tight_layout()

# Show the combined plot
plt.show()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
import matplotlib.pyplot as plt

# Specify the file name for the PDF
pdf_file_name = 'Car plots.pdf'

# Save plots to a PDF file
with PdfPages(pdf_file_name) as pdf:
    for fig_num in plt.get_fignums():
        fig = plt.figure(fig_num)  # Get the figure by its number
        pdf.savefig(fig)  # Save the current figure to the PDF file

# Output success message
print(f"Plots saved to {pdf_file_name}")
```
